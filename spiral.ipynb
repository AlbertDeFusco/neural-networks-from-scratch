```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scratchnet.datasets import make_spiral

X, y = make_spiral(n_samples=600, n_features=2, n_classes=2)

plt.scatter(X[:,0], X[0:,1], c=y, cmap=plt.cm.Spectral)
```

```python
from sklearn.svm import SVC
from plots import decision_boundary

svc = SVC(C=10)
svc.fit(X, y)

decision_boundary(svc, X, y)
svc.score(X, y)
```

```python
from scratchnet.networks import Network
from scratchnet.losses import CrossEntropy
from scratchnet.activations import Tanh, Sigmoid, Relu
```

```python
single_layer = Network(2)
single_layer.add_layer(128, Relu(), random_state=42)
single_layer.add_layer(128, Relu(), random_state=42)
single_layer.add_layer(1, Sigmoid(), random_state=42)
```

```python
history = single_layer.train(X, y, CrossEntropy(), epochs=10000)
```

```python
print(single_layer.score(X,y))

fig, axes = plt.subplots(ncols=2, figsize=(16,6))

axes[0].plot(history)
decision_boundary(single_layer, X, y, ax=axes[1])
```
